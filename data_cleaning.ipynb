### Importing necessary library and loading the dataset.

```python
import pandas as pd
```

```python
dt = pd.read_excel("data_penjualan.xlsx")
dt.head()
```

### Adding some new values into column

```python
dt.loc[dt['REFEREE'].str.contains('Makassar', case=False), 'CABANG'] = 'MAKASSAR'
dt.loc[dt['REFEREE'].str.contains('Surabaya', case=False), 'CABANG'] = 'SURABAYA'
dt.CABANG.unique()
```

### Dropping unnecessary columns

```python
dt = dt[dt['TEST_INV_GROUP_NAME'] != 'BANK DARAH']
dt = dt[dt['CABANG'] != 'BANDUNG']
dt.TEST_INV_GROUP_NAME.unique()
```

### Add new column "TARGET_SALES"

```python
dt["TARGET_SALES"] = None
dt
```
